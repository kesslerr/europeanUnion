# european Union

This is just a short exploration of the data of "parteivergleich.eu" to see how similar different political partys are in their positions to a particular set of topics (prior to election of european parliament 2019).
No guarantee for the correctness of the data.

```python
import numpy as np
import pandas as pd; pd.set_option('display.max_columns', 100)
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set(color_codes=True)
```

```python
df = pd.read_excel("parteivergleich_eu.xlsx")
df = df.replace(to_replace={"Ja":   1,
                      "Nein": -1,
                      #"k.A.": np.nan, # find a solution/workaround for this
                      "k.A.": 0,
                      np.nan: 0})
df.columns = df.columns.str.strip()
df
```

```python
g = sns.clustermap(df.drop(["Themenblock","Frage"], axis = 1))
```

## fancy up the plot

```python
# row colors (themenblock)
lutr = dict(zip(df["Themenblock"].unique(), sns.color_palette("tab20", n_colors=len(df["Themenblock"].unique()), desat=.5)))
row_colors = df["Themenblock"].map(lutr)

# column colors (party color)
lutc = {"Grüne*": "green",
       "AfD": "blue",
       "CDU*": "black",
       "CSU*": "black",
       "FDP": "yellow",
       "SPD": "red",
       "Die Linke": "orange",
       "Die Humanisten": "purple"}
col_colors = df.columns[2:].map(lutc)
col_colors = col_colors.fillna( value = "white" ) #.to_series()
```

```python
g = sns.clustermap(df.drop(["Themenblock","Frage"], axis = 1),
                   row_colors=row_colors, col_colors=col_colors,
                   row_cluster = False, col_cluster = True,
                  )
# Draw the legend bar for the topics                 
for label in lutr.keys():
    g.ax_col_dendrogram.bar(0, 0, color=lutr[label],
                            label=label, linewidth=0)
g.ax_col_dendrogram.legend(bbox_to_anchor=(0, -1)) ##loc="upper right", ncol=1)

# Adjust the postion of the main colorbar for the heatmap
g.cax.set_position([.98, .2, .03, .45])
```

# now visualize similarity per Topic !

```python
for topic in df["Themenblock"].unique():
    df2 = df[df["Themenblock"]==topic]
    g = sns.clustermap(df2.drop(["Themenblock","Frage"], axis = 1),
                      col_colors=col_colors,
                      row_cluster = False, col_cluster = True,
                      figsize=(10,3))
    plt.title(topic)
    plt.show()
```
